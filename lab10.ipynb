```python
import cv2 as cv
import numpy as np 
from matplotlib import pyplot as plt
```

## load coins image/convert to binary

```python
src = cv.imread("1.png")
gray_img = cv.cvtColor(src,cv.COLOR_BGR2GRAY)
_,bin_img = cv.threshold(gray_img,127,255,cv.THRESH_BINARY)
plt.imshow(bin_img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
```

```python
rect11x11 = cv.getStructuringElement(cv.MORPH_RECT,(11,11))
cross11x11 = cv.getStructuringElement(cv.MORPH_CROSS,(11,11))
ellipse11x11 = cv.getStructuringElement(cv.MORPH_ELLIPSE,(11,11))
erode_rect11x11_img = cv.erode(bin_img, rect11x11, iterations = 4)
erode_cross11x11_img = cv.erode(bin_img, cross11x11, iterations = 4)
erode_ellipse11x11_img = cv.erode(bin_img,ellipse11x11, iterations = 4)
print("Original Binary")
plt.imshow(bin_img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
print("Result from dilated_rect11x11")
plt.imshow(erode_rect11x11_img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
print("Result from dilated_cross11x11")
plt.imshow(erode_cross11x11_img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()
print("Result from dilated_ellipse11x11")
plt.imshow(erode_ellipse11x11_img,cmap='gray'),plt.xticks([]),plt.yticks([])
plt.show()

```
